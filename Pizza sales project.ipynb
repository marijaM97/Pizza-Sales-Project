```python
# Convert date (order_date) & time (order_time) to proper datetime format

df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y')
df['order_time'] = pd.to_datetime(df['order_time'], format='%H:%M:%S')
```

```python
# Adding a column for the hour
df['hour'] = df['order_time'].dt.hour 

# Adding a column for the month
df['month'] = df['order_date'].dt.month 

# Adding a column for the week
df['week'] = df['order_date'].dt.isocalendar().week  
```

```python
# Total Revenue

import pandas as pd
import ipywidgets as widgets
from ipywidgets import interactive

df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y', errors='coerce')

def total_revenue_by_date_range(date_range):
    start_date, end_date = date_range
    filtered_data = df[(df['order_date'] >= start_date) & (df['order_date'] <= end_date)]
    
    total_revenue = filtered_data['total_price'].sum()
    
    print(f"Date Range: {start_date.date()} to {end_date.date()} - Total Revenue: ${total_revenue:.2f}")

start_date = pd.Timestamp('2015-01-01')
end_date = pd.Timestamp('2015-12-31')

date_range_slider = widgets.SelectionRangeSlider(
    options=pd.date_range(start=start_date, end=end_date).tolist(),
    index=(0, len(pd.date_range(start=start_date, end=end_date)) - 1),
    description='Date Range:',
    orientation='horizontal',
    layout={'width': '80%'},
    style={'description_width': 'initial'},
)

interactive_plot = interactive(total_revenue_by_date_range, date_range=date_range_slider)

display(interactive_plot)

```

```python
# Average Order Value

import pandas as pd
import ipywidgets as widgets
from ipywidgets import interactive

df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y', errors='coerce')

def average_order_value_by_date_range(date_range):
    start_date, end_date = date_range
    filtered_data = df[(df['order_date'] >= start_date) & (df['order_date'] <= end_date)]
    
    total_revenue = filtered_data['total_price'].sum()
    
    total_orders = len(filtered_data)
    
    if total_orders > 0:
        aov = total_revenue / total_orders
    else:
        aov = 0 
    
    print(f"Date Range: {start_date.date()} to {end_date.date()}")
    print(f"Average Order Value (AOV): ${aov:.2f}\n")

start_date = pd.Timestamp('2015-01-01')
end_date = pd.Timestamp('2015-12-31')

date_range_slider = widgets.SelectionRangeSlider(
    options=pd.date_range(start=start_date, end=end_date).tolist(),
    index=(0, len(pd.date_range(start=start_date, end=end_date)) - 1),
    description='Date Range:',
    orientation='horizontal',
    layout={'width': '80%'},
    style={'description_width': 'initial'},
)

interactive_plot = interactive(average_order_value_by_date_range, date_range=date_range_slider)

display(interactive_plot)
```

```python
# Monthly Trend for Total Orders

import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interactive

df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y', errors='coerce')

def plot_monthly_trend(date_range):
    start_date, end_date = date_range
    filtered_data = df[(df['order_date'] >= start_date) & (df['order_date'] <= end_date)]
    
    filtered_data['month'] = filtered_data['order_date'].dt.month
    
    monthly_orders = filtered_data.groupby('month')['order_id'].nunique()
    
    plt.figure(figsize=(12, 6))
    monthly_orders.plot(kind='line', marker='o', color='teal')
    plt.title(f'Monthly Trend for Total Orders', fontsize=16)
    plt.xlabel('Month of the Year', fontsize=14)
    plt.ylabel('Total Orders', fontsize=14)
    plt.grid(True)
    plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    plt.show()

start_date = pd.Timestamp('2015-01-01')
end_date = pd.Timestamp('2015-12-31')

date_range_slider = widgets.SelectionRangeSlider(
    options=pd.date_range(start=start_date, end=end_date).tolist(),
    index=(0, len(pd.date_range(start=start_date, end=end_date)) - 1),
    description='Date Range:',
    orientation='horizontal',
    layout={'width': '80%'},
    style={'description_width': 'initial'},
)

interactive_plot = interactive(plot_monthly_trend, date_range=date_range_slider)

display(interactive_plot)
```

```python
# Percentage of Sales by Pizza Category

import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interactive

df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y', errors='coerce')

def sales_by_category_date_range(start_date, end_date):
    start_date = pd.to_datetime(start_date).date()
    end_date = pd.to_datetime(end_date).date()

    
    filtered_df = df[(df['order_date'].dt.date >= start_date) & (df['order_date'].dt.date <= end_date)]
    
  

    if len(filtered_df) == 0:
        print("Nema podataka za dati period.")
    
    sales_by_category = filtered_df.groupby('pizza_category')['total_price'].sum()

    sales_by_category.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), colormap='Set2')
    plt.title(f'Percentage of Sales by Pizza Category from {start_date.strftime("%B %d, %Y")} to {end_date.strftime("%B %d, %Y")}')
    plt.ylabel('')
    plt.show()

start_date_picker = widgets.DatePicker(
    description='Start Date',
    value=pd.to_datetime('2015-01-01').date(),  
    disabled=False
)

end_date_picker = widgets.DatePicker(
    description='End Date',
    value=pd.to_datetime('2015-12-31').date(),  
    disabled=False
)

interactive_plot = interactive(sales_by_category_date_range, start_date=start_date_picker, end_date=end_date_picker)

display(interactive_plot)



```

```python
# Total Pizzas Sold by Pizza Category

import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from ipywidgets import interact


def total_pizzas_sold_by_category_date_range(start_date, end_date):
    start_date = pd.to_datetime(start_date)
    end_date = pd.to_datetime(end_date)
    
    filtered_df = df[(df['order_date'] >= start_date) & (df['order_date'] <= end_date)]
    
    total_sold_by_category = filtered_df.groupby('pizza_category')['quantity'].sum().reset_index()
    
    total_sold_by_category = total_sold_by_category.sort_values(by='quantity', ascending=False)

    fig = px.funnel(total_sold_by_category, x='quantity', y='pizza_category', title='Total Pizzas Sold by Pizza Category')
    
    fig.show()

start_date_widget = widgets.DatePicker(
    description='Start Date', 
    value=pd.to_datetime('2015-01-01'),
    min=pd.to_datetime('2015-01-01'),
    max=pd.to_datetime('2015-12-31')
)

end_date_widget = widgets.DatePicker(
    description='End Date', 
    value=pd.to_datetime('2015-12-31'),
    min=pd.to_datetime('2015-01-01'),
    max=pd.to_datetime('2015-12-31')
)

interact(total_pizzas_sold_by_category_date_range, start_date=start_date_widget, end_date=end_date_widget)

```

```python
# Top 5 Best Sellers

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interactive

df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y', errors='coerce')

def plot_top_5_sales_by_date_range(start_date, end_date):
    filtered_data = df[(df['order_date'] >= start_date) & (df['order_date'] <= end_date)]

    top_5_revenue = filtered_data.groupby('pizza_name')['total_price'].sum().nlargest(5)

    top_5_quantity = filtered_data.groupby('pizza_name')['quantity'].sum().nlargest(5)

    top_5_orders = filtered_data['pizza_name'].value_counts().nlargest(5)
    sns.set(style="whitegrid")

    plt.figure(figsize=(14, 12))

    # Top 5 Best Sellers by Revenue
    plt.subplot(3, 1, 1)
    sns.barplot(x=top_5_revenue.index, y=top_5_revenue.values, hue=top_5_revenue.index, palette='Blues_d', legend=False)
    plt.title('Top 5 Best Sellers by Revenue', fontsize=16)
    plt.xlabel('Pizza Name', fontsize=12)
    plt.ylabel('Revenue', fontsize=12)
    for i in range(len(top_5_revenue)):
        plt.text(i, top_5_revenue.values[i] + 5, f'{top_5_revenue.values[i]:.2f}', ha='center', va='bottom', fontsize=10)

    # Top 5 Best Sellers by Quantity
    plt.subplot(3, 1, 2)
    sns.barplot(x=top_5_quantity.index, y=top_5_quantity.values, hue=top_5_quantity.index, palette='Greens_d', legend=False)
    plt.title('Top 5 Best Sellers by Quantity', fontsize=16)
    plt.xlabel('Pizza Name', fontsize=12)
    plt.ylabel('Quantity', fontsize=12)
    for i in range(len(top_5_quantity)):
        plt.text(i, top_5_quantity.values[i] + 1, f'{top_5_quantity.values[i]}', ha='center', va='bottom', fontsize=10)

    # Top 5 Best Sellers by Orders
    plt.subplot(3, 1, 3)
    sns.barplot(x=top_5_orders.index, y=top_5_orders.values, hue=top_5_orders.index, palette='Reds_d', legend=False)
    plt.title('Top 5 Best Sellers by Orders', fontsize=16)
    plt.xlabel('Pizza Name', fontsize=12)
    plt.ylabel('Number of Orders', fontsize=12)
    for i in range(len(top_5_orders)):
        plt.text(i, top_5_orders.values[i] + 1, f'{top_5_orders.values[i]}', ha='center', va='bottom', fontsize=10)

    plt.tight_layout()
    plt.show()

# Definisanje widgeta za izbor datuma
start_date_widget = widgets.DatePicker(
    description='Start Date',
    value=pd.to_datetime('2015-01-01')
)

end_date_widget = widgets.DatePicker(
    description='End Date',
    value=pd.to_datetime('2015-12-31')
)

# Aktiviranje interaktivnog prikaza sa funkcijom koja filtrira podatke na osnovu datuma
interactive_plot = interactive(plot_top_5_sales_by_date_range, start_date=start_date_widget, end_date=end_date_widget)
output = interactive_plot.children[-1]
output.layout.height = '600px'

# Prikazivanje interaktivnog prikaza
display(interactive_plot)


```
